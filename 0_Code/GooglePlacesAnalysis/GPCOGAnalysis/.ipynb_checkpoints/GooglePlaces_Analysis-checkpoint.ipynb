```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from PIL import Image
```

```python
with open('venue_data.json') as f:
    venue_data = json.load(f)
```

```python
#extract keys from data for iteration
keys = []
for key in venue_data.keys():
    keys.append(key)
```

```python
#pull relevant info from google data
venue_lat = []
venue_lng = []
venue_name = []
venue_id = []
venue_types = []
for i in range(len(venue_data)):
    lat = venue_data[keys[i]]['lat']
    lng = venue_data[keys[i]]['long']
    name = venue_data[keys[i]]['name']
    GoogleID = venue_data[keys[i]]['id']
    venue_type = str(venue_data[keys[i]]['type'])
    
    venue_lat.append(lat)
    venue_lng.append(lng)
    venue_name.append(name)
    venue_id.append(GoogleID)
    venue_types.append(venue_type)
```

```python
#create new dataframe
df = pd.DataFrame()
df['name'] = venue_name
df['lat'] = venue_lat
df['long'] = venue_lng
df['Google_ID'] = venue_id
df['Venue_Type'] = venue_types

df.head()
```

```python
#using places from google data search for available popular times data
venues_pop_times = []
for i in range(len(df)):
    id = df.loc[i]['Google_ID']
    a = populartimes.get_id(google_key, id)
    venues_pop_times.append(a)
    #print(((i+1)/(len(venues)))*100, "% complete")
```

```python
#inspect first item
venues_pop_times[0]
```

```python
#only extract places that have pop times data
venues_final = []
for i in range(len(venues_pop_times)):
    try:
        if venues_pop_times[i]['populartimes']:
            venues_final.append(venues_pop_times[i])
    except KeyError:
        pass
print(len(venues_final),"venues have populartimes data")
```

```python
#add relevent data to datframe
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
for day in range(7):
    df[days[day]] = 0
    for venue in venues_final:
        name = venue['name']
        dayOfWeek = venue['populartimes'][day]['name']
        count = sum(venue['populartimes'][day]['data'])
        lat = venue['coordinates']['lat']
        lng = venue['coordinates']['lng']
        idx = venue['id']
        index = df[df['Google_ID'] == idx].index.values[0]
        
        df.at[index, days[day]] = count
        
        '''print("Venue: ", name)
        print("Day of Week: ", dayOfWeek)
        print("Lat: ", lat)
        print("Long: ", lng)
        print("Visitor Count: ", count)
        print(df[df['Google_ID'] == idx])
        print(df[df['Google_ID'] == idx].index.values[0])
        print("----------------------------------------------------")'''
df.head(10)
```

```python
#sum the week's data for each venue
df['WeekTotals'] = 0
for i in range(len(df)):
    a = []
    for j in range(7):
        count = df.at[i, days[j]]
        a.append(count)
        #print(a)
    df.at[i, 'WeekTotals'] = sum(a)
df.head(10)
```

```python
#tokenize venue types
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

for i in range(len(df)):
    string = df.loc[i]['Venue_Type']
    tokens = tokenizer.tokenize(string)
    [token.lower() for token in tokens]
    df.at[i,'Venue_Type'] = tokens
df.head(10)
```

```python
#get unique list of venue types
list_of_venue_types = []
for i in range(len(df)):
    a = df.iloc[i]['Venue_Type']
    for venue in a:
        #print(venue)
        venue = venue.lower()
        list_of_venue_types.append(venue)
list_of_venue_types = list(set(list_of_venue_types))
list_of_venue_types
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(vocabulary=list_of_venue_types)

r = pd.SparseDataFrame(cv.fit_transform(df['Venue_Type']), 
                       df.index,
                       cv.get_feature_names(), 
                       default_fill_value=0)
```

```python
df_encoded = pd.concat([df, r], axis = 1)
df_encoded
```

```python
df_train = df_encoded.drop(['Venue_Type','Google_ID','name','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],axis=1)
df_train.head()
```

```python
df_train['lat'] = df_train['lat']/max(df_train['lat'])
df_train.head()
```

```python
max_lat = max(df_train['lat'])
min_lat = min(df_train['lat'])
max_lng = max(df_train['long'])
min_lng = min(df_train['long'])

def normalize_lat(lat, max, min):
    norm_lat = (lat - min) / (max - min)
    return norm_lat

def normalize_lng(lng, max, min):
    norm_lng = (lng - min) / (max - min)
    return norm_lng

for i in range(len(df_train)):
    lat = df_train.iloc[i]['lat']
    lng = df_train.iloc[i]['long']
    norm_lat = normalize_lat(lat,max_lat,min_lat)
    norm_lng = normalize_lng(lng,max_lng,min_lng)
    df_train.at[i,'lat'] = norm_lat
    df_train.at[i,'long'] = norm_lng
df_train.head()
```

```python
X = df_train[df_train['WeekTotals']>0].drop(['WeekTotals'], axis=1)
y = df_train[df_train['WeekTotals']>0]['WeekTotals']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np.array(X), np.array(y), test_size=0.1, random_state=42)
```

```python
print("X_train:",len(X_train),"samples")
print("y_train:",len(y_train),"samples")
print("X_test:",len(X_test),"samples")
print("y_test:",len(y_test),"samples")
print(X_train.shape)
```

## Train MLP

```python
from keras.models import Sequential
from keras.layers import Conv2D, Dense, BatchNormalization, Flatten
from keras.optimizers import Adam
from keras.layers import MaxPooling2D, Dropout

#build 1D MLP
model = Sequential()
model.add(Dense(38, activation='relu', input_shape = (len(X_train.columns),)))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

checkpointer = ModelCheckpoint(filepath='model.weights.best.hdf5', verbose=1, save_best_only=True)
earlystopper = EarlyStopping(monitor='val_loss', min_delta=0.0001, patience=3, verbose=1, mode='auto')

hist = model.fit(X_train, y_train, epochs=100, 
                 validation_data=(X_test, y_test),
                 callbacks=[checkpointer, earlystopper],
                 steps_per_epoch = 1000, validation_steps=10)
```

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy

base_model = RandomForestRegressor(n_estimators = 10, random_state = 42)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)

best_random = rf_random.best_estimator_
random_accuracy = evaluate(best_random, X_test, y_test)

print('Improvement of {:0.2f}%.'.format( 100 * (random_accuracy - base_accuracy) / base_accuracy))
```
